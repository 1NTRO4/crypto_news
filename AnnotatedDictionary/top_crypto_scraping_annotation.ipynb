```python
## for first 50 exchanges
import requests
from bs4 import BeautifulSoup

def get_top_cryptocurrencies():

    url = 'https://coinranking.com/exchanges?page=1' ## for companies
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table containing the top cryptocurrencies
    table = soup.find('table')

    top_cryptocurrencies = []
    # sym = []

    # Iterate over the rows in the table
    for row in table.find_all('tr'):  # Skip the header row
        # Get the name of the exchange company
        name_cell = row.find('a', {'class': 'profile__link'})  ## for name of crypto and also for company

        if name_cell:
            name = name_cell.get_text(strip=True)

            top_cryptocurrencies.append(name.lower())


        # Stop after getting 50 cryptocurrencies
        if len(top_cryptocurrencies) >= 50:
            break

    return top_cryptocurrencies


# Get the top 200 cryptocurrencies
top_crypto_ex1 = get_top_cryptocurrencies()

print(top_crypto_ex1)
```
